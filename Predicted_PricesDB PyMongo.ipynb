```python
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
mongo = MongoClient(port=27017)
```

```python
db = mongo['predicted_prices']
```

```python
print(db.list_collection_names())
```

```python
predicted = db['predicted_pricesDB']
```

```python
predicted.count_documents({})
```

```python
print(db.predicted_pricesDB.find_one())
```

```python
query = [{'$group': {'_id': "$manufacturer", 'count': { '$sum': 1 }}}]
```

```python
results = list(predicted.aggregate(query))
```

```python
print("Number of manufacturers in result: ", len(results))
```

```python
pprint(results[0:38])
```

```python
results_df = pd.DataFrame(results)
results_df.head(10)
```
